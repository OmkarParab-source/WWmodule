```python
import os
import pandas as pd
import numpy as np
from pydub import AudioSegment
from scipy.io import wavfile
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_info(file):
    rate, data = wavfile.read(file)
    return rate, data
```

```python
def get_spectrogram(file):
    rate, data = get_info(file)
    nfft = 250
    fs = 800
    noverlap = 125
    pxx, freqs, bins, im = plt.specgram(data, nfft, fs, noverlap = noverlap)
    return pxx
```

```python
# Testing

def load_tests():
    positives = []
    negatives = []
    backgrounds = []
    for file in os.listdir("./tests"):
        if "positive" in file:
            # audiofile = AudioSegment.from_wav("./tests/"+file)
            positives.append("./tests/"+file)
        elif "negative" in file:
            # audiofile = AudioSegment.from_wav("./tests/"+file)
            negatives.append("./tests/"+file)
        if "background" in file:
            # audiofile = AudioSegment.from_wav("./tests/"+file)
            backgrounds.append("./tests/"+file)
    return positives, negatives, backgrounds
```

```python
(positives, negatives, backgrounds) = load_tests()
```
